This code showcase how easy is to develope a model to forecast a time seris, by using PyCaret 3.0.

The aim of this is not to provide the best possibile model, but a better way to manage an ML project.

For more infor read my article on Medium.

```python
import pandas as pd
import numpy as np
from pycaret.time_series import TSForecastingExperiment, get_config, set_current_experiment, models, predict_model
import os
```

```python
air_pass = pd.read_csv('https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv')
print('Shape of the data= ', air_pass.shape)
print('Column datatypes= \n',air_pass.dtypes)
air_pass
```

```python
air_pass['Month'] = pd.to_datetime(air_pass.Month)
air_pass = air_pass.set_index(air_pass.Month)
air_pass.drop(columns=['Month'], inplace = True)
print('Column datatypes= \n',air_pass.dtypes)
air_pass
```

```python
fh = 2 #forecast horizon indicates how many points in the future we want to forecast
experiment = TSForecastingExperiment()
experiment.setup(data=air_pass, fh=fh, target='Passengers', session_id=42)
```

```python
experiment.check_stats()
```

```python
metric = 'mape'
n_select = 1
best_model = experiment.compare_models(sort=metric,  n_select = n_select)
```

```python
experiment.predict_model(best_model)
```
